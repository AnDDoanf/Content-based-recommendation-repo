```python
import numpy as np
import pandas as pd
```

```python
credits_df = pd.read_csv("credits.csv", error_bad_lines=False, engine="python")
movies_df = pd.read_csv('movies.csv', error_bad_lines=False, engine="python")
```

```python
movies_df = movies_df.merge(credits_df, on="title")
movies_df.info()
```

Determine features that are in considering

```python
movies_df = movies_df[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df.dropna(inplace=True)
movies_df.isnull().sum()
```

```python
movies_df.duplicated().sum()
```

```python
movies_df.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
  lst = []
  for i in ast.literal_eval(obj):
    lst.append(i["name"])
  return lst

def convertt(obj):
  lst = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 1:
      lst.append(i['name'])
      counter += 1
    else:
      break
  return lst

def converttt(obj):
  lst = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      lst.append(i["name"])
  return lst
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df['cast'] = movies_df['cast'].apply(convertt)
movies_df['crew'] = movies_df['crew'].apply(converttt)
movies_df.head()
```

```python
movies_df["overview"] = movies_df["overview"].apply(lambda x: x.split())
movies_df.head()
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
movies_df.head()
```

```python
movies_df["tags"] = movies_df["overview"] + movies_df["genres"] + movies_df["keywords"] + movies_df["cast"] + movies_df["crew"]
movies = movies_df[["movie_id", "title", "tags"]]
movies.head()
```

```python
import string
```

```python
movies["tags"] = movies["tags"].apply(lambda x: ' '.join(x))
movies["tags"] = movies["tags"].apply(lambda x: x.lower())
movies["tags"] = movies["tags"].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
vectors = cv.fit_transform(movies["tags"]).toarray()
```

```python
vectors.shape
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies["tags"] = movies["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
def recommend(movie, movies):
  movie_index = movies[movies['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
  movies_list = [movies.iloc[i[0]].title for i in movies_list]
  return movies_list
```

```python
recommend("Iron Man", movies)
```
